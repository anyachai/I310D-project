```python
import pandas as pd
import matplotlib.pyplot as plt

col_list = ["RELIG16", "HAPPY", "DIVORCE", "CLASS"]
df = pd.read_csv("GSS2018.csv", usecols=col_list)
```

```python
df.dropna(subset = ["CLASS"], inplace=True)
df.dropna(subset = ["DIVORCE"], inplace=True)
df.dropna(subset = ["HAPPY"], inplace=True)
df.dropna(subset = ["RELIG16"], inplace=True)
df.drop(df.index[df["RELIG16"] == "DK"], inplace=True)

print(df.to_string())
```

```python
df.sample(n=300)
print(df.to_string())
```

```python
count = df['CLASS'].value_counts()
plt.figure();
count.plot.bar();
```
